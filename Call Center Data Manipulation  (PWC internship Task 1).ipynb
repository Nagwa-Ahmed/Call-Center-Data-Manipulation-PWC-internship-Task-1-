```python
import pandas as pd
```

```python
df=pd.read_excel('Call Center.xlsx')
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df1=df['Date'].str.split('-',expand=True)
```

```python
df['year']=df1[0]
df['month']=df1[1]
df['day']=df1[2]
df.drop(columns=['Date'],inplace=True)
```

```python
df.head()
```

```python
df['Time']=df['Time'].astype(str)
df['AvgTalkDuration']=df['AvgTalkDuration'].astype(str)
```

```python
df2=df['AvgTalkDuration'].str.split(':',expand=True)
```

```python
df['minutes_talk_duration']=df2[1]
df['seconds_talk_duration']=df2[2]
df.drop(columns=['AvgTalkDuration'],inplace=True)
```

```python
df3=df['Time'].str.split(':',expand=True)
```

```python
df['Hour']=df3[0]
df['Minute']=df3[1]
df.drop(columns=['Time'],inplace=True)
```

```python
df.to_excel('Call Center(answered& not).xlsx',index=False)
```
